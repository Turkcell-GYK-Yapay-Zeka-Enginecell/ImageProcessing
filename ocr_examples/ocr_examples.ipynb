```python
import easyocr

reader = easyocr.Reader(['tr'])  # Trke iin
result = reader.readtext('Fatura.webp', detail=0)

# Metni birletir
text = "\n".join(result)

# Dosyaya yaz
with open('fatura_metni.txt', 'w', encoding='utf-8') as f:
    f.write(text)
    print(text)

print("Metin 'fatura_metni.txt' dosyasna kaydedildi.")
```
