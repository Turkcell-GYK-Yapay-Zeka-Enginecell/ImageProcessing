```python
import sys
import os
from PyQt5.QtWidgets import (
    QApplication, QWidget, QPushButton, QLabel,
    QFileDialog, QVBoxLayout, QMessageBox
)
from PyQt5.QtGui import QPixmap
from rembg import remove
```

```python
class BackgroundRemover(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Arka Plan Kaldrc")
        self.setGeometry(100, 100, 500, 600)

        self.layout = QVBoxLayout()

        self.label = QLabel(" Bir resim sein", self)
        self.label.setStyleSheet("font-size: 16px; padding: 10px;")
        self.layout.addWidget(self.label)

        self.image_label = QLabel(self)
        self.layout.addWidget(self.image_label)

        self.select_button = QPushButton(" Resim Se", self)
        self.select_button.clicked.connect(self.select_image)
        self.layout.addWidget(self.select_button)

        self.remove_button = QPushButton(" Arka Plan Kaldr", self)
        self.remove_button.clicked.connect(self.remove_background)
        self.remove_button.setEnabled(False)
        self.layout.addWidget(self.remove_button)

        self.result_label = QLabel(" Sonu:", self)
        self.layout.addWidget(self.result_label)
        self.result_image = QLabel(self)
        self.layout.addWidget(self.result_image)

        self.setLayout(self.layout)
        self.image_path = None
        self.output_path = None

    def select_image(self):
        file_name, _ = QFileDialog.getOpenFileName(self, "Resim Se", "", "Images (*.png *.jpg *.jpeg *.bmp *.tiff)")
        if file_name:
            self.image_path = file_name
            self.image_label.setPixmap(QPixmap(self.image_path).scaledToWidth(300))
            self.remove_button.setEnabled(True)

    def remove_background(self):
        if not self.image_path:
            return

        try:
            with open(self.image_path, 'rb') as input_file:
                input_data = input_file.read()
                output_data = remove(input_data)

            base_name = os.path.splitext(self.image_path)[0]
            self.output_path = f"{base_name}_no_bg.png"

            with open(self.output_path, 'wb') as output_file:
                output_file.write(output_data)

            # Sonu grselini gster
            self.result_image.setPixmap(QPixmap(self.output_path).scaledToWidth(300))

            QMessageBox.information(self, "Baarl", f"Arka plan kaldrld:\n{self.output_path}")
        except Exception as e:
            QMessageBox.critical(self, "Hata", f"lem srasnda hata olutu:\n{e}")
```

```python
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = BackgroundRemover()
    window.show()
    sys.exit(app.exec_())
```
